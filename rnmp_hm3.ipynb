```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score
import joblib
```

```python
from sklearn.linear_model import LogisticRegression

models = {
    'RandomForest': RandomForestClassifier(),
    'GradientBoosting': GradientBoostingClassifier(),
    'LogisticRegression': LogisticRegression()
}

param_grids = {
    'RandomForest': {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]},
    'GradientBoosting': {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 1.0]},
    'LogisticRegression': {'C': [0.01, 0.1, 10], 'penalty': ['l1', 'l2']}
}

best_models = {}
for name, model in models.items():
    print(f"Training {name}...")
    grid_search = GridSearchCV(model, param_grids[name], cv=3, scoring='f1')
    grid_search.fit(offline.drop(columns=['Diabetes_binary']), offline['Diabetes_binary'])
    best_models[name] = grid_search.best_estimator_
    print(f"Best parameters for {name}: {grid_search.best_params_}")

best_f1_score = 0
best_model_name = None
for name, model in best_models.items():
    y_pred = model.predict(online.drop(columns=['Diabetes_binary']))
    f1 = f1_score(online['Diabetes_binary'], y_pred)
    print(f"F1 Score for {name}: {f1}")
    if f1 > best_f1_score:
        best_f1_score = f1
        best_model_name = name

best_model = best_models[best_model_name]
print(best_model)
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import GBTClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator


assembler = VectorAssembler(
    inputCols=[col for col in offline_df.columns if col != 'Diabetes_binary'],
    outputCol="features"
)

offline_transformed = assembler.transform(offline_df)
online_transformed = assembler.transform(online_df)

gbt = GBTClassifier(labelCol="Diabetes_binary", featuresCol="features", maxIter=200, stepSize=0.1)
gbt_model = gbt.fit(offline_transformed)

predictions = gbt_model.transform(online_transformed)
evaluator = BinaryClassificationEvaluator(labelCol="Diabetes_binary")
accuracy = evaluator.evaluate(predictions)

print("Accuracy:", accuracy)

model_path = "gradient_boosting_model"
gbt_model.save(model_path)

spark.stop()
```
